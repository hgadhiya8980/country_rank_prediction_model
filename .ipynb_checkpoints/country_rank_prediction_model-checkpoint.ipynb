```python
X = df1.drop("Rank", axis=1)
y = df1["Rank"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
lr.score(X_test, y_test)
```

```python
rd = Ridge()
rd.fit(X_train, y_train)
y_pred1 = rd.predict(X_test)
rd.score(X_test, y_test)
```

```python
lc = Lasso()
lc.fit(X_train, y_train)
y_pred2 = lc.predict(X_test)
lc.score(X_test, y_test)
```

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(X_train, y_train)
y_pred3 = svr.predict(X_test)
svr.score(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)
y_pred4 = dtr.predict(X_test)
dtr.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)
y_pred5 = rfr.predict(X_test)
rfr.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
cv = cross_val_score(dtr, X_train, y_train, cv=10)
cv, cv.mean()
```

```python
def country_rank_prediction(model,Name, City, State, category, Student_Population, Total_Annual_Cost):
    for num,var in enumerate(value):
        if var == Name:
            Name = num
    for num,var in enumerate(value1):
        if var == City:
            City = num
    for num,var in enumerate(value2):
        if var == State:
            State = num
    for num,var in enumerate(value3):
        if var == category:
            category = num
    x = np.zeros(len(X.columns))
    x[0] = Name
    x[1] = City
    x[2] = State
    x[3] = category
    x[4] = Student_Population
    x[5] = Total_Annual_Cost
    
    x = sc.transform([x])[0]
    return model.predict([x])[0]
```
